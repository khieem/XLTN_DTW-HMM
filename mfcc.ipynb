## Đọc âm thanh

```python
aa, _     = librosa.load(pa)
ab, _     = librosa.load(pb)
alen, _   = librosa.load(plen)
axuong, _ = librosa.load(pxuong)
atrai, _  = librosa.load(ptrai)
aphai, _  = librosa.load(pphai)
aban, _   = librosa.load(pban)
```

## Trích xuất đặc trưng MFCC, delta, deltadelta

### MFCC

```python
mfcc_a     = librosa.feature.mfcc(y=aa)
mfcc_b     = librosa.feature.mfcc(y=ab)
mfcc_len   = librosa.feature.mfcc(y=alen)
mfcc_xuong = librosa.feature.mfcc(y=axuong)
mfcc_trai  = librosa.feature.mfcc(y=atrai)
mfcc_phai  = librosa.feature.mfcc(y=aphai)
mfcc_ban   = librosa.feature.mfcc(y=aban)
```

```python
mfcc_a.shape
```

```python
librosa.display.specshow(mfcc_a, x_axis="time")
plt.colorbar(format="%2.f")
plt.show()
```

### delta

```python
delta_a     = librosa.feature.delta(mfcc_a)
delta_b     = librosa.feature.delta(mfcc_b)
delta_len   = librosa.feature.delta(mfcc_len)
delta_xuong = librosa.feature.delta(mfcc_xuong)
delta_trai  = librosa.feature.delta(mfcc_trai)
delta_phai  = librosa.feature.delta(mfcc_phai)
delta_ban   = librosa.feature.delta(mfcc_ban)
```

```python
librosa.display.specshow(delta_a, x_axis="time")
plt.colorbar(format="%+2.f")
plt.show()
```

### deltadelta

```python
delta2_a     = librosa.feature.delta(mfcc_a, order=2)
delta2_b     = librosa.feature.delta(mfcc_b, order=2)
delta2_len   = librosa.feature.delta(mfcc_len, order=2)
delta2_xuong = librosa.feature.delta(mfcc_xuong, order=2)
delta2_trai  = librosa.feature.delta(mfcc_trai, order=2)
delta2_phai  = librosa.feature.delta(mfcc_phai, order=2)
delta2_ban   = librosa.feature.delta(mfcc_ban, order=2)
```

```python
librosa.display.specshow(delta2_a, x_axis="time")
plt.colorbar(format="%+2.f")
plt.show()
```
